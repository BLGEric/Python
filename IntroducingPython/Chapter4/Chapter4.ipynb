```python
#Lambda
def edit_story(words, func):
    for word in words:
        print(func(word))
```

```python
stairs = ['哈囉', '你好', '嘿', '最近如何']
```

```python
def enliven(word):
    return word.capitalize() + '!'
```

```python
edit_story(stairs, enliven)
```

```python
edit_story(stairs, lambda word: word.capitalize() + '!')
```

```python
def my_range(first=0, last=10, step=2):
    number = first
    while number < last:
        yield number #yield 簡單說來就是一個生成器，生成器是這樣一個函數，它記住上一次返回時在函數體中的位置。
        number += step
```

```python
ranger = my_range(1, 9)
```

```python
for x in ranger:
    print(x)
```
